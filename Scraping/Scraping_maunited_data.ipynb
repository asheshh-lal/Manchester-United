### Webscraping for Manchester United Data

```python
import requests
import pandas as pd
```

```python
data = requests.get("https://fbref.com/en/squads/19538871/2022-2023/all_comps/Manchester-United-Stats-All-Competitions")

standard_stats = pd.read_html(data.text, match='Standard Stats')[0]
```
